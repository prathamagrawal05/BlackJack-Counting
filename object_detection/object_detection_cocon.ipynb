```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model,config_file)
```

```python
class_labels = []
filename='labels.txt'
with open(filename,'rt') as fpt:
    class_labels = fpt.read().rstrip('\n').split('\n')
```

```python
print(class_labels)
```

```python
print(len(class_labels))
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5) ##255/2 = 1217.5
model.setInputMean((127.5,127.5,127.5))
model.setInputSwapRB(True)
```

# Read Image


```python
img = cv2.imread('car.jpg')
```

```python
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ClassIndex, confidece, bbox = model.detect(img,confThreshold=0.6)
```

```python
print(ClassIndex)
```

```python
font_scale =3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img,boxes,(255,0,0), 2 )
    cv2.putText(img,class_labels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, color = (0,255,0),thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
cap = cv2.VideoCapture(1)

#check if itn is open correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcame")
    
font_scale =3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret,frame = cap.read()
    
    ClassIndex, confidece, bbox = model.detect(frame, confThreshold=0.6)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if(ClassInd<=80):
                cv2.rectangle(frame,boxes,(255, 0, 0), 2)
                cv2.putText(frame,class_labels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, color = (0, 255,0), thickness = 3)
    
    cv2.imshow('Object Detection Tutorial', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
        
cap.realease()
cv2.destroyAlWindows()
```
